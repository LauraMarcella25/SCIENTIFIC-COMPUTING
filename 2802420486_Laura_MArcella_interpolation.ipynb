```python
from scipy.interpolate import interp1d
import matplotlib.pyplot as plt

```

```python
x = [0, 1, 2]
y = [1, 3, 2]

f = interp1d(x, y)

y_hat = f(1.5)
print(y_hat)
print(plt.style.available)
```

```python
plt.figure(figsize=(10, 9)) #ukuran plot 10 x 8 inci
plt.plot(x, y, '-ob') #menampilkan titik data dengan garis biru dan lingkaran
plt.plot(1.5, y_hat, 'ro') #menampilan titik hasil interpolasi dengan warna merah
plt.title('Linear interpolation at x = 1.5') #judul grafik
plt.xlabel('x') #label sumbu x
plt.ylabel('y') #label sumbu y
plt.show() #menampilkan grafik


```

```python
import numpy as np
import matplotlib.pyplot as plt

def divided_diff(x, y):
    n = len(y)
    coef = np.zeros([n, n])  # Membuat tabel kosong untuk divided differences
    coef[:, 0] = y  # Kolom pertama berisi nilai y

    for j in range(1, n):  # Loop untuk mengisi kolom-kolom berikutnya
        for i in range(n - j):
            coef[i][j] = (coef[i + 1][j - 1] - coef[i][j - 1]) / (x[i + j] - x[i])

    return coef
def newton_poly(coef, x_data, x):
    n = len(x_data) - 1
    p = coef[n]

    for k in range(1, n + 1):
        p = coef[n - k] + (x - x_data[n - k]) * p

    return p
```

```python
x = np.array([-5, -1, 0, 2])
y = np.array([-2, 6, 1, 3])

# Menghitung koefisien divided difference
a_s = divided_diff(x, y)[0, :]

# Evaluasi interpolasi pada titik-titik baru
x_new = np.arange(-5, 2.1, .1)
y_new = newton_poly(a_s, x, x_new)

# Membuat plot
plt.figure(figsize=(12, 8))
plt.plot(x, y, 'bo')  # Menampilkan titik data asli
plt.plot(x_new, y_new)  # Menampilkan kurva interpolasi
```

```python
from scipy.interpolate import CubicSpline
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('seaborn-v0_8-poster')

x = [0, 1, 2]
y = [1, 3, 2]

# use bc type = 'natural' adds the constraints as we described above
f = CubicSpline(x, y, bc_type='natural')
x_new = np.linspace(0, 2, 100)
y_new = f(x_new)


plt.figure(figsize=(10, 8))
plt.plot(x_new, y_new, 'b')  # Garis interpolasi
plt.plot(x, y, 'ro')  # Titik data asli
plt.title('Cubic Spline Interpolation')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

distances = np.array([0, 5, 10])
speeds = np.array([10, 15, 20])

target_distance = 7.2
estimated_speed = np.interp(target_distance, distances, speeds)

print(f"Kecepatan lari pada jarak {target_distance} meter adalah {estimated_speed:.2f} m/detik")

x_new = np.linspace(0, 10, 100)
y_new = np.interp(x_new, distances, speeds)


plt.plot(x_new, y_new, label="Interpolasi", color="blue")
plt.scatter(distances, speeds, color="red", label="Data Asli")

plt.xlabel("Jarak (meter)")
plt.ylabel("Kecepatan (m/detik)")
plt.title("Interpolasi Kecepatan Lari Berdasarkan Jarak")
plt.show()


```
